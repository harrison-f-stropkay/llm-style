```python
import pandas as pd

authors_df = pd.read_csv("../intermediate_data/authors.csv")
temps_df = pd.read_csv("../intermediate_data/temps.csv")

# drop rows in temps_df that have author=="reddit"
temps_df = temps_df[temps_df.author != "reddit"]
combined_df = pd.concat([authors_df, temps_df], axis=0, ignore_index=True)

combined_df.to_csv("../data.csv", index=False)

combined_df
```
